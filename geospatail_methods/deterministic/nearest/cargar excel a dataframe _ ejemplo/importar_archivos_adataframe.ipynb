```python

import pandas as pd

ruta = "pvt_data_uno.xlsx"

todas_hojas = pd.read_excel(
    ruta,
    sheet_name=None,   
    skiprows=3,
    usecols="B:G"
)

# Limpiar cada hoja y poner ID como índice
pvt_dict = {}

for nombre_hoja, df in todas_hojas.items():
    # Nos aseguramos de que ID sea índice
    df_limpio = df.set_index("ID")

    # Normalizamos el nombre de la hoja para usarlo como clave/variable
    nombre_clave = nombre_hoja.strip().replace(" ", "_").lower()
    pvt_dict[nombre_clave] = df_limpio

# Verifica
for nombre, df in pvt_dict.items():
    print(nombre, df.shape)
    
#vizualicacion 
pvt_dict["pakistan"]

```
